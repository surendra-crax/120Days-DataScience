```python
import pandas as pd
df = pd.read_csv("titanic.csv")
print(df.groupby("Sex")["Age"].mean())
```

```python
print(df.groupby("Pclass")["Survived"].mean())
```

```python
#grouping multiple columns
print(df.groupby(["Sex","Pclass"])["Survived"].mean())
```

```python
print(df.groupby("Pclass")["PassengerId"].count())
print(df["Pclass"].value_counts())
```

```python
#aggregation set
print(df.groupby("Pclass").agg({
    "Age":"mean",
    "Fare":"mean",
    "Survived":"mean"
}))
```

Merging is just joining two tables using a common column.
Like SQL joins.

You will use this in:

cleaning

combining datasets

mapping domain work

data preprocessing

```python
#merging dataframes
df1 = pd.DataFrame({
    "PassengerId": [1, 2, 3, 4],
    "Name": ["A", "B", "C", "D"],
    "Age": [22, 34, 29, 40]
})
print(df1)
```

```python
df2 = pd.DataFrame({
    "PassengerId": [2, 3, 4, 5],
    "Fare": [71.28, 8.05, 13.00, 25.50],
    "Pclass": [1, 3, 3, 2]
})


print(df2)
```

```python
merged = pd.merge(df1,df2, on="PassengerId")
print(merged)
```

**✔ 1. A “key column” is required**

Example: PassengerId, CustomerID, RollNo
That column must be present in both tables.

**✔ 2. Merge joins two DataFrames **

Just like VLOOKUP (Excel) or JOIN (SQL).

**✔ 3. Types of joins**

Only remember these:

- inner → keep matching rows

- left → keep all left rows (most used)

- outer → keep everything

That’s all the theory you need to do real work.

```python
#inner join
import pandas as pd
pd.merge(df1,df2,on="PassengerId",how="inner")
```

```python
pd.merge(df1, df2, on="PassengerId", how="left")
```

```python
pd.merge(df1, df2, on="PassengerId", how="outer")
```

# PIVOT TABLES

A pivot table is just a summary table that organizes data in two dimensions:

✔ Rows (index)

✔ Columns

✔ Values (what to calculate)

What does a pivot table do?

It answers questions like:

- What is average fare for each class?

- What is survival rate for male vs female?

- What is survival rate for each class for each gender?

- What is the mean age for each embark point?

- It is powerful because it summarizes thousands of rows into a clean table.

```python
#Survival rate by Class
df.pivot_table(
    index="Pclass",
    values="Survived",
    aggfunc="mean"
)
```

```python
#mean Age rate by Gender
df.pivot_table(
    index="Sex",
    values = "Age",
    aggfunc="mean"
)
```

```python
#survival rate by sex
df.pivot_table(
    index="Sex",
    values="Survived",
    aggfunc="mean"
)
```

```python
#survival rate by class and sex
df.pivot_table(
    index="Sex",
    columns="Pclass",
    values="Survived",
    aggfunc="mean"
)
```

```python
#Average Fare by Embarked & Class
df.pivot_table(
    index="Embarked",
    columns="Pclass",
    values="Fare",
    aggfunc="mean"
)
```

```python
#counting how many passengers in each class
df.pivot_table(index="Pclass", values="Fare", aggfunc="count")
```

```python
#maximum fare per class
df.pivot_table(index="Pclass", values="Fare", aggfunc="max")
```

```python
#practice tasks
#value_counts()
print(df["Sex"].value_counts())
print(df["Pclass"].value_counts())
print(df["Embarked"].value_counts())
```

```python
#Multi-column groupby
print(df.groupby(["Sex", "Pclass"])["Fare"].mean())
print(df.groupby(["Embarked", "Pclass"])["Survived"].mean())
```

```python
#Aggregation with agg()
df.groupby("Pclass").agg({
    "Age":"mean",
    "Fare":"mean",
    "Survived":"mean"
})
```

```python
#Different merge types
inner = pd.merge(df1, df2, on="PassengerId", how="inner")
left = pd.merge(df1, df2, on="PassengerId", how="left")
outer = pd.merge(df1, df2, on="PassengerId", how="outer")
print(inner)
print(left)
print(outer)
```

```python
#Pivot table variations
print(df.pivot_table(index="Pclass", columns="Sex", values="Age", aggfunc="mean"))

print(df.pivot_table(index="Embarked", values="Fare", aggfunc=["mean", "max"]))
```
