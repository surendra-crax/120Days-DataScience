```python
#simple calculator
x = int(input("Enter your first number: "))
y = int(input("Enter your second number: "))

print("Select your operation")
print("1.Addition")
print("2.Subtraction")
print("3.Multiplication")
print("4.Division")

entry = input("Enter any number(1/2/3/4):")

if entry == "1":
  print("Result:",x+y)
elif entry == "2":
  print("Result:",x-y)
elif entry == "3":
  print("Result:",x*y)
elif entry == "4":
  print("Result:",x/y)
else:
  print("Invalid operation")
```

```python
#Temperature convertor

print("Select your conversion")
print("1.Celsius-->Fahrenheit")
print("2.Fahrenheit-->Celsius")

entry = input("Enter your choice(1/2): ")


if entry == "1":
  C = float(input("Enter temperature in celsius: "))
  res = (C * 9/5) + 32
  print("Result:",res,"F")
elif entry == "2":
  F = float(input("Enter temperature in fahrenheit: "))
  res = (F - 32) * 5/9
  print("Result:",res,"C")
else:
  print("Invalid Choice")
```

```python
#Reverse using indexing
str1 = "Hello"
lol = len(str1) - 1
revStr = ""
for x in range(lol,-1,-1):
  revStr = revStr+str1[x]
print(revStr)
```

```python
#Reverse String Using a Backward Loop
str1="Hello"
lol = len(str1)-1
revStr= ""
while lol >= 0:
  revStr= revStr+ str1[lol]
  lol-=1

print(revStr)
```

```python
#revese using slicing
str1= "Hello"
revStr = str1[::-1]
print(revStr)
```

```python
#Mutliplication table
x = int(input("Enter the table you want: "))

for i in range(1,11):
  res= x * i
  print(x , "*",i,"=",res)
```

```python
#Sum of first N numbers
N = int(input("Enter the number"))
sum = 0
for x in range(1,N+1):
  sum = sum + x

print("Result",sum)
```
