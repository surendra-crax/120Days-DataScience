# While loop examples

```python
#Print numbers from 1 to 10 using a while loop
i=1
while i<11:
  print(i)
  i +=1
```

```python
#Reverse a number using while loop (e.g., 123 → 321)
x = int(input("Enter the number to reverse"))
rev = 0
while x>0:
  digit = x%10
  rev = (rev * 10)+digit
  x = x//10

print(rev)
```

```python
#prime number checking using optimized square root method
x = int(input("Enter the number:"))
y = int(x**0.5)+1
if x <=1:
  print("Not a prime")
else:
  for i in range(2,y):
    if x%i==0:
      print("Not a prime")
      break
  else:
      print("Prime")
```

```python
#prime number checking using optimized square root method using math import
import math as m
x = int(input("Enter the number:"))
y = int(m.sqrt(x))
if x <=1:
  print("Not a prime")
else:
  for i in range(2,y+1):
    if x%i==0:
      print("Not a prime")
      break
  else:
      print("Prime")
```

```python
#factorial of number
x = int(input("Enter a number :"))
result=1
if x==0:
  print("Factorial of this number is 1")
elif x==1:
  print("Factorial of number is 1")
else:
  for i in range(1,x+1):
    result=result*i
  print("Factorial of a number is :",result)
```

```python
#pyramid printing
x = int(input("Enter the number of rows : "))

for row in range(1,x+1):
  #printing spaces
  for space in range(1,x-row+1):
    print(" ",end="")

  #printing stars
  for star in range(1,2*row):
    print("*",end="")

  print()
```

```python
#inverted pyramid
x = int(input("Enter the number of rows : "))

for row in range(1,x+1):
  #printing spaces
  for space in range(row-1):
    print(" ",end="")

  #printing stars
  for star in range(2*(x-row+1)-1):
    print("*",end="")

  print()
```

```python
#inverted printing
x=int(input("Enter the number of rows: "))

for row in range(1,x+1):
  #printing spaces
  for spaces in range(row-1):
    print(" ",end="")

  #printing numbers
  for num in range(1,2*(x-row+1)-1):
    print(num,end="")

  print()
```
