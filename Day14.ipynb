#List comprehensions
1️⃣ What is a List Comprehension?

A list comprehension is a shorter and cleaner way to create a list using a loop.

It does exactly the same work as a normal for loop that appends values to a list.

Nothing new logically.
Only syntax changes.

```python
#normal loop
result = []
for i in range(1,6):
  result.append(i*1)
```

```python
print(result)
```

```python
#list comprehension syntax
result1 = [i*1 for i in range(1,6)]
print(result1)
```

1. for i in range(1, 6) → loop

2. i * i → value added to the list

So Python:

- loops over numbers

- squares each

- stores result automatically

No append() needed.

```python
#list comprehension wit condition
result3 = [i*1 for i in range(1,11) if i%2==0]
print(result3)
```

```python
result4 = ["Even" if i%2 ==0 else "Odd" for i in range(1,6)]
print(result4)
```

```python
#Numbers from 1 to 10
num = [i for i in range(1,11)]
print(num)
```

```python
#cubes of numbers from 1 to 5
num1 = [i**3 for i in range(1,6)]
print(num1)
```

```python
#only odd numbers form 1 to 20

num2 = [i for i in range(1,21) if i%2!=0]
print(num2)
```

```python
#length of each word in sentence
sentence = "Python makes coding simple"
resultt = [len(word) for word in sentence.split()]
print(resultt)
```

```python
#convert list of strings to lowercase
words = ["PYTHON","Data","SCIENCE"]
result12 = [word.lower() for word in words]
print(result12)
```

#Dictionary comprehension
What is it?

A dictionary comprehension creates a dictionary using a loop, in one line.

```python
squares = {i:i*i for i in range(1,6)}
print(squares)
```

```python
#filtering dictionary values
data = {"a":10,"b":3,"c":7,"d":15}
filtered = {k: v for k,v in data.items() if v>5}
print(filtered)
```

```python
#create a dictionary mapping numbers 1-5 to their cubes
cubes = {i:i**3  for i in range(1,6)}
print(cubes)
```

```python
#filter a dictionary to keep only even values
Items = {"a":2,"b":3,"c":7,"d":10}
even_filters = {k:v for k,v in Items.items() if v %2==0}
print(even_filters)
```

```python
#create a dictionary mapping words
sentence = "Python makes coding simple"
word_length = {word:len(word) for word in sentence.split()}
print(word_length)
```

#Set comprehension
What is it?

A set comprehension creates a set (unique values only).

```python
#unique squares
nums = [1,2,2,3,3,3,2,2,3,4,4,5,6,7,5,5]
unique_squares = {i**2 for i in nums}
print(unique_squares)
```

```python
#filtering even numbers
nums1 = {1,2,3,4,5,6}
evens = {i for i in nums if i% 2 == 0}
print(evens)
```

```python
#create a set of unique vowels from a string
string = "Yaswanth is very innocent and intelligent"
unique_vowels = {i.lower() for i in string if i.lower() in "aeiou"}
print(unique_vowels)
```

```python
#create a set of numberd divisible by 3 from 1 to 20
div_by_3 = {i for i in range(1,21) if i%3==0}
print(div_by_3)
```

#Lambda Function
A lambda is a one-line function without a name.

```python
def square(x):
  return x*x
square(2)
```

```python
square=lambda x:x*x
square(5)
```

```python
#lambda with map()
nums = [1,2,3,4]
result = list(map(lambda x : x*x,nums))
print(result)
```

```python
#lambda with filter()
nums = [1,2,3,4,5,6]
even = list(filter(lambda x:x%2==0,nums))
print(evens)
```

```python
#use lambda to find square of a number
square = lambda x:x*x
square(5)
```

```python
#use map() to convert list of strings to uppercase
words = ["Yashwanth","Kishore","Rajesh"]
str_uppercase = list(map(lambda x:x.upper(),words))
print(str_uppercase)
```

```python
#use filter() to keep numbers greater than 10
nums = [12,4,5,333,4,5,4,33,3,44,55,5]
greater_10 = list(filter(lambda x:x>10,nums))
print(greater_10)
```

```python
#use lambda to check if number is even

evens = lambda x:x%2==0
print(evens(2))
```

```python
#use map() to add 5 to every number in a list
nums =[1,2,3,4,5,6]
added = list(map(lambda x : x+5,nums))
print(added)
```

#sorted() with CUSTOM key FUNCTION

```python
nums =[1,4,2,3,6,7,4,5]
print(sorted(nums))
```

```python
#sorting list of tuples
data =[(1,3),(4,1),(2,5)]
result = sorted(data,key=lambda x:x[1])
print(result)
```

```python
#sorting in descending order
result = sorted(data,key=lambda x:x[1],reverse=True)
print(result)
```

```python
#sort list of dictionaries
students =[
    {"name":"A","marks":85},
    {"name":"B","marks":72},
    {"name":"C","marks":90}
]
result = sorted(students,key=lambda x:x["marks"])
print(result)
```

```python
result = sorted(students,key=lambda x:x["marks"],reverse=True)
print(result)
```

```python
#sorting strings by length
words = ["python","ai","machine","data"]
result = sorted(words,key=lambda x:len(x))
print(result)
```

```python
#sort list of tuples by second value
data1 = [(1,8),(8,5),(4,6)]
result = sorted(data1,key=lambda x :x[1])
print(result)
```

```python
#sort list of dictionaries by a numeric field
students =[
    {"name":"A","salary":50000},
    {"name":"B","salary":70000},
    {"name":"C","salary":30000}
]

result = sorted(students,key=lambda x:x["salary"])
print(result)
```

```python
#sort words by length
words = ["python","ai","machine","data"]

sorted_words = sorted(words,key=lambda x:len(x))
print(sorted_words)
```

```python
#sort number by absolute value
nums = [-10,3,-2,7,-5]

sorted_nums = sorted(nums,key=lambda x:abs(x))
print(sorted_nums)
```

```python
#sort students by marks in descending order
students =[
    {"name":"A","marks":85},
    {"name":"B","marks":72},
    {"name":"C","marks":90}
]
result = sorted(students,key=lambda x:x["marks"],reverse=True)
print(result)
```
