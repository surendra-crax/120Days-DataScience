#PROBLEM STATEMENT

You are given student data in the form of a list of dictionaries.

Each student has:

- name

- marks

Your task is to analyze this data using Python intermediate features.

```python
students = [
    {"name": "Ravi", "marks": 85},
    {"name": "Anu", "marks": 72},
    {"name": "Kiran", "marks": 90},
    {"name": "Sita", "marks": 60},
    {"name": "Arjun", "marks": 45}
]
```

```python
#extracting all marks
marks = [s["marks"] for s in students]
print(marks)
```

```python
#filter passed students marks>=50
passed_students = list(filter(lambda s: s["marks"] >= 50, students))
print(passed_students)
```

```python
#adding grade to each student >=80(A),>=60(B) else(C)

graded_students= [
    {
        "name":s["name"],
        "marks":s["marks"],
        "grade":"A" if s["marks"] >=80 else "B" if s["marks"] >=60 else "C"
    }
    for s in students
]
print(graded_students)
```

```python
#sort students by marks
ranked_students = sorted(students, key=lambda s: s["marks"], reverse=True)
print(ranked_students)
```

```python
#create name:marks diectory
student_dict = {s["name"]:s["marks"] for s in students}
print(student_dict)
```

```python
#gettings topper name
topper = max(students,key=lambda s:s["marks"])
print(topper)
```
