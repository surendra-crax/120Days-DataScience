# WHAT IS PANDAS? (Real Meaning)

Pandas is a Python library used to work with tabular data — the same type of data you see in:

- CSV files
- Excel sheets
- Database tables
- Mapping datasets
- Business reports
- Logs, metrics, structured text

If you can understand Pandas, you can analyze any structured data.

# WHY head(), tail(), info(), describe() MATTER

These four methods are your “eyes”.

**✔ head()**

Shows first 5 rows.
Used to quickly inspect if data loaded correctly.

**✔ tail()**

Shows last 5 rows.
Used to check formatting, missing values at bottom.

**✔ info()**

Shows:

- number of rows

- number of columns

- datatype of each column

- missing values

This is extremely important for data cleaning.

**✔ describe()**

Gives summary statistics for numeric columns:

- mean

- min

- max

- std (standard deviation)

- quartiles

This helps you understand data distribution instantly.

# Missing Data (isnull, dropna, fillna)

````python
print(df.isnull().sum())
````

````python
#selecting rows with missing data
print(df[df["Age"].isnull()])
````

````python
#missing any columns
print(df[df.isnull().any(axis=1)])
````

````python
#removing missing rows
print(df.dropna())
````

````python
print(df.dropna(subset=["Age"]))
````

````python
#fill missing values
df["Age"].fillna(df["Age"].mean(),inplace=True)
print(df.tail())
````

````python
df["Cabin"].fillna("Unknown",inplace=True)
````

| Task                     | Command                   |
| ------------------------ | ------------------------- |
| Check missing            | `df.isnull().sum()`       |
| Select rows with missing | `df[df["Age"].isnull()]`  |
| Drop missing rows        | `df.dropna()`             |
| Fill missing values      | `df["Age"].fillna(value)` |
