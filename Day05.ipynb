# Error Handling
**The try block lets you test a block of code for errors.**

**The except block lets you handle the error.**

**The finally block lets you execute code, regardless of the result of the try- and except blocks.**

```python
try:
  print(x)
except:
  print("An exception occured")
```

```python
#DIVISION BY ERROR
try:
    a = 10
    b = 0
    result = a / b
    print(result)

except ZeroDivisionError:
    print("Error: You tried to divide by zero.")
```

```python
#File-Not-Found
try:
  f = open("file.txt","r")
  content = f.read()
  print(content)
except FileNotFoundError:
  print("Error: File does not exist.")
```

```python
#Invalid inut handler
try:
  user_input= int(input("Enter a number: "))
  print("You entered",user_input)

except ValueError:
  print("Error: Please enter a valid integer.")
```

```python
#index out of range handler
try:
  lst= [10,20,30]
  print(lst[5])
except IndexError:
  print("Error: Index is out of range")
```

```python
#dictionary keyerror handler
try:
  student = {"name":"surendra", "age":23}
  print(student["city"])
except:
  print("Error: Key not found in dictionary.")
```

```python
#TypeErro Handler
try:
  result = "Age: "+ 21 #invalid combinatin str and int
except TypeError:
  print("Error: Cannot add string and integer. ")
```

```python
#custom exceptions
class AgeError(Exception):
     pass

age= -5
if age<0:
  raise AgeError("Age cannot be negative")
```

```python
#catching custom error) try-except
try:
  age= -3
  if age<0:
    raise AgeError("Age cannot be negative")
except AgeError as e:
  print("custom Error: ",e)
```

```python
#custom error -2--password strength
class WeakPassword(Exception):
  pass

password= "abc"

try:
  if len(password) < 6 :
    raise WeakPassword("Password too short")

except WeakPassword as e:
  print("Custom error: ",e)
```

```python
#invalid score
class InvalidScore(Exception):
  pass

try:
  score= 120
  if score>100:
    raise InvalidScore("Score cannnot be grater than 100")
except InvalidScore as e:
  print("Custom Error:",e)
```
