# Functions
A function is a reusable block of code that performs one task.

```python
def greet():
  print("Hello")

greet()
greet()
```

```python
#Functions with parameters
def add(a,b):       # a,b are parameters
  print(a+b)

add(1,2)            # 1,2 are arguments
```

```python
# return keyoword in functions --> return sends back a value from the functions
def square(n):
  return n*n
def add(a,b):
  return a+b

print(square(5))
print(add(5,6))
```

```python
#default arguments
def greet(name="user"):
  print("Hello",name)

greet("Surya")
```

```python
# *args--> used when you dont know how many arguments youll get
def total(*args):
  print(args)
  return sum(args)

print(total(1,2,3,4,45))
```

```python
# **kwargs --> used for allowing passing any number of key-value pairs
def user_info(**kwargs):
  for key,value in kwargs.items():
    print(key, "=",value)
user_info(Name ="Surendra",age = 21, city = "Rjy")
```

```python
def student_info(*args,**kwargs):
  #print(args,kwargs)
  #print(kwargs)
  return args,kwargs


student_info("Surendra" , "Maths" , Name = "surya", age="28",classroom = "10")
```

```python
#factorial
def factorial(n):
  if n ==0 or n==1:
    return 1

  result=1

  for i in range(2,n+1):
    result = result*i

  return result

factorial(4)
```

```python
# recursuve version fo rfactorial
def factorial(n):
  if n==0 or n==1:
    return 1

  return n*factorial(n-1)
factorial(5)
```

```python
#reversing string using functions
def rev_str(s):
  result = ""

  for i in range(len(s)-1,-1,-1):
    result= result+s[i]

  print(result)

rev_str("Hello")
```

```python
# vowels counter from a sentence
def count_vowels(c):
  vowels = ['a','e','i','o','u']
  count = 0

  for char in c:
    if char.lower() in vowels:
      count = count+1

  return count

sentence = "He is very brave and intelligent "
count_vowels(sentence)
```

```python
# findin glaregst from list
def largest_num(lst):

  largest = lst[0]

  for i in lst:
    if i > largest:
      largest = i

  return largest

lst = [1,5,4,6,88,9,33,4,5,6,887,6,7,6,8,453,234,5,45,65,75,678]
largest_num(lst)
```

```python
#*args multiplier
def multiply_all(*args):
  result = 1
  for i in args:
    result = result*i
  return result

print(multiply_all(1,2,3,4))
print(multiply_all(-1,3,-2))
```

```python
# duplicates removal
def rem_dup(lst):

  uq_lst = []

  for i in lst:
    if i not in uq_lst:
      uq_lst.append(i)

  return uq_lst

lst = [1,3,2,4,5,2,1,5,6,7,8,87,6,4,34]
rem_dup(lst)
```

# File Handling

The key function for working with files in Python is the open() function.

The open() function takes two parameters; filename, and mode.

There are four different methods (modes) for opening a file:

**"r" - Read - Default value. Opens a file for reading, error if the file does not exist**

**"a" - Append - Opens a file for appending, creates the file if it does not exist**

**"w" - Write - Opens a file for writing, creates the file if it does not exist**

**"x" - Create - Creates the specified file, returns an error if the file exists**

```python
with open("sample.txt", "w") as f:
  f.write(" Hello! Welcome to demofile.txt \n This file is for testing purposes. \n Good Luck!")
```

```python
#remove punctuation function
import string

def remove_punctuation(text):
  cleaned = ""
  for char in text:
    if char not in string.punctuation:
      cleaned+= char
  return cleaned

text = "Hello!!! This, is... Data-Science."
print(remove_punctuation(text))
```

```python
#clean text function
def clean_text(text):
  #lowercase
  text = text.lower()

  #removing punctuations
  cleaned=""
  for char in text:
    if char not in string.punctuation:
      cleaned +=char

  #removing extra spaces
  cleaned = " ".join(cleaned.split())

  return cleaned

text = "Hello!!!   THIS is   DATA-Science, 2025..."
print(clean_text(text))
```
